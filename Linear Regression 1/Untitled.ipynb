```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('linear_regression.csv')
df.head()
```

```python
X=df[['X']]
y=df['Y']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
df.isna()
```

```python
lreg=LinearRegression()
lreg.fit(X_train,y_train)
```

```python
lreg_pred_y=lreg.predict(X_test)
print(lreg_pred_y)
```

```python
train_score=lreg.score(X_train,y_train)
test_score=lreg.score(X_test,y_test)
print(f"Train Score: {train_score}")
print(f"Test Score: {test_score}")

train_pred = lreg.predict(X_train)
test_mse = mean_squared_error(y_test, lreg_pred_y)
train_mse = mean_squared_error(y_train, train_pred)

print(f"Train MSE: {train_mse}")
print(f"Test MSE: {test_mse}")
```

```python
x_range = np.linspace(X_test.min().values, X_test.max().values, 100).reshape(-1, 1)
y_range = lreg.predict(x_range)

plt.scatter(X_test, y_test, color='blue', label="Dataset Points")
plt.plot(x_range, y_range, color='red', label="Regression Line")

plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
